```python
print("___ Import Module ___")
import pandas as pd
import os
from matplotlib import pyplot as plt
from lifelines import KaplanMeierFitter


os.chdir('./T-Matsuura_LDLT')
path = os.getcwd()
```

```python
train_y_180 = pd.read_csv(path+'/Data/train_y_180.csv',index_col = 0)
val_Y_180 = pd.read_csv(path+'/Data/val_Y_180.csv',index_col = 0)
survival_analysis = pd.concat([train_y_180,val_Y_180])
survival_analysis = survival_analysis[survival_analysis['graftloss'] == 1]
c = survival_analysis['graft_days_aive']
d = survival_analysis['graftloss']
```

```python
#survival analysis

fig, ax = plt.subplots(figsize=(12,12))#figure size
ax.set_facecolor('white')

kmf_1=KaplanMeierFitter()
kmf_1.fit(c,d,label='Patients with graft loss')
kmf_1.plot_survival_function(show_censors    = True)

#detail
plt.ylim(0,1)
plt.yticks([0.0,0.25,0.5,0.75,1.0],fontsize=50,weight='normal')
plt.xticks([180,2000,4000,6000],fontsize=50,weight='normal')
plt.legend().set_visible(False)
plt.xlabel('Postoperative days',weight='normal',size = 48)
plt.ylabel('Survival Probability',weight='normal',size = 48)
ax.axvline(x=180, color='black')
ax.spines['top'].set_color('white')
ax.spines['right'].set_color('white')
plt.rcParams["font.family"] = "Arial" #font
plt.savefig(path+'/Output/Fig1/Fig1A.png',bbox_inches='tight')
plt.show()

```
